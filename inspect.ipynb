```python

import json, os

# 차수 → 시험일자 매핑
CHA_SU_MAP = {
    0: '2026년 3월 18일 (0차)',
    1: '2026년 4월 8일 (1차)',
    2: '2026년 7월 24일 (2차)',
}

cols2 = ['사번', '이름', '직위', '직급', '본부', '사업부', '담당', '조직',
         '지필점수', '지필면제여부', '차수', '기타', '비고']
df2 = df[cols2].copy()
df2['사번'] = df2['사번'].astype(str).str.strip().str.split('.').str[0]
df2 = df2.fillna('')
df2['지필점수'] = df2['지필점수'].astype(str).str.strip()
df2.loc[df2['지필점수'].isin(['nan','None','']), '지필점수'] = '미입력'
df2['기타'] = df2['기타'].astype(str).str.strip()
df2.loc[df2['기타'].isin(['nan','None','']), '기타'] = ''
df2['비고'] = df2['비고'].astype(str).str.strip()
df2.loc[df2['비고'].isin(['nan','None','']), '비고'] = ''

data_dict2 = {}
for _, row in df2.iterrows():
    sabun = row['사번']
    if not sabun or sabun == 'nan':
        continue
    cha_su_num = row['차수']
    try:
        cha_su_num = int(float(cha_su_num)) if cha_su_num != '' else -1
    except:
        cha_su_num = -1
    exam_date = CHA_SU_MAP.get(cha_su_num, '미정')

    entry = {
        '이름': row['이름'],
        '직위': row['직위'],
        '직급': row['직급'],
        '본부': row['본부'],
        '사업부': row['사업부'],
        '담당': row['담당'],
        '조직': row['조직'],
        '지필점수': row['지필점수'],
        '지필면제여부': row['지필면제여부'],
        '시험일자': exam_date,
        '기타': row['기타'],
        '비고': row['비고'],
    }
    # 동일 사번이 여러 차수인 경우 최신(큰 차수) 기준으로 덮어쓰기
    if sabun in data_dict2:
        prev_cha = data_dict2[sabun].get('_cha', -1)
        if cha_su_num >= prev_cha:
            entry['_cha'] = cha_su_num
            data_dict2[sabun] = entry
    else:
        entry['_cha'] = cha_su_num
        data_dict2[sabun] = entry

# _cha 키 제거
for v in data_dict2.values():
    v.pop('_cha', None)

print(f"총 {len(data_dict2)}명")
print("샘플:", list(data_dict2.items())[:1])
```

```python

import json, os

CHA_SU_MAP = {
    0: '2026년 3월 18일 (0차)',
    1: '2026년 4월 8일 (1차)',
    2: '2026년 7월 24일 (2차)',
}

cols2 = ['사번', '이름', '직위', '직급', '본부', '사업부', '담당', '조직',
         '지필점수', '지필면제여부', '차수', '기타', '비고', '이메일']
df2 = df[cols2].copy()
df2['사번'] = df2['사번'].astype(str).str.strip().str.split('.').str[0]
df2 = df2.fillna('')
df2['지필점수'] = df2['지필점수'].astype(str).str.strip()
df2.loc[df2['지필점수'].isin(['nan','None','']), '지필점수'] = '미입력'
df2['기타'] = df2['기타'].astype(str).str.strip()
df2.loc[df2['기타'].isin(['nan','None','']), '기타'] = ''
df2['비고'] = df2['비고'].astype(str).str.strip()
df2.loc[df2['비고'].isin(['nan','None','']), '비고'] = ''
df2['이메일'] = df2['이메일'].astype(str).str.strip()
df2.loc[df2['이메일'].isin(['nan','None','']), '이메일'] = ''

# 사번별 리스트로 집계 (차수 오름차순)
data_multi = {}
for _, row in df2.iterrows():
    sabun = row['사번']
    if not sabun or sabun == 'nan':
        continue
    try:
        cha_num = int(float(row['차수'])) if row['차수'] != '' else -1
    except:
        cha_num = -1
    entry = {
        '이름': row['이름'],
        '직위': row['직위'],
        '직급': row['직급'],
        '본부': row['본부'],
        '사업부': row['사업부'],
        '담당': row['담당'],
        '조직': row['조직'],
        '지필점수': row['지필점수'],
        '지필면제여부': row['지필면제여부'],
        '시험일자': CHA_SU_MAP.get(cha_num, '미정'),
        '차수': cha_num,
        '기타': row['기타'],
        '비고': row['비고'],
        '이메일': row['이메일'],
    }
    if sabun not in data_multi:
        data_multi[sabun] = []
    data_multi[sabun].append(entry)

# 차수 오름차순 정렬
for k in data_multi:
    data_multi[k].sort(key=lambda x: x['차수'])

# 중복 있는 사번 확인
dups = {k: v for k, v in data_multi.items() if len(v) > 1}
print(f"여러 차수에 걸친 사번 수: {len(dups)}")
for k, v in dups.items():
    print(f"  {k}: {v[0]['이름']} - 차수 {[e['차수'] for e in v]}")
```
